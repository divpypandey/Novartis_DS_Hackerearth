First Import all the essential libraries like numpy,pandas


```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Importing the Train dataset and for analysis I just need to find how many columns have missing values.
Note: There is no need for date columns so I just removed it from dataset.

```
dataset = pd.read_csv('sample_data/Train.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
# below line is shown how many columns have how many missing value
dataset.isnull().sum()
```

For filling the missing value Scikit-learn hase SimpleImputer library that fill the values according to the mean of the columns.

```
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(missing_values=np.nan,strategy='mean')
imputer.fit(X[:,:])
X[:,:]=imputer.transform(X[:,:])
```

I split the dataset into 25% to 75% ratio

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)

```

First I tried Logistic Regression but it did not give fruitful answers So I just used Xgboost that lead me to 99.51% accuracy

```
from xgboost import XGBClassifier
xgb_model = XGBClassifier(max_depth=2, n_estimators=7600).fit(X_train,y_train)
prediction = xgb_model.predict(X_test)
```
